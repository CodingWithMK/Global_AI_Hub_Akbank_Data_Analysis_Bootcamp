```python
def checking_column_types(dataframe: pd.DataFrame, categorical_treshold: int=10, cardinal_treshold: int=20):
  categorical_columns = [column for column in dataframe.columns if dataframe[column].dtype == '0']

  numerical_but_categorical = [column for column in dataframe.columns if dataframe[column].nunique() < categorical_treshold and dataframe[column].dtypes != '0']

  categorical_but_cardinal = [column for column in dataframe.columns if dataframe[column].nunique() > cardinal_treshold and dataframe[column].dtypes == '0']

  categorical_columns = [column for column in categorical_columns if column not in categorical_but_cardinal]

  # Numerical Columns
  numerical_columns = [column for column in dataframe.columns if dataframe[column].dtypes != '0']
  numerical_columns = [column for column in numerical_columns if column not in numerical_but_categorical]

  # Columns Types and Overview
  print("--------------------- Categorical Columns ---------------------")
  print(categorical_columns)

  print('\n')

  print("--------------------- Numerical Columns ---------------------")
  print(numerical_columns)

  print('\n')

  print("--------------------- Categorical but Cardinal Columns ---------------------")
  print(categorical_but_cardinal)

  print('\n')

  print("--------------------- Numerical but Categorical Columns ---------------------")
  print(numerical_but_categorical)

  print('\n')

  print("--------------------- Columns Info ---------------------")
  print(f"Observations: {dataframe.shape[0]}")
  print(f"Variables: {dataframe.shape[1]}")
  print(f"Categorical Columns: {len(categorical_columns)}")
  print(f"Numerical Columns: {len(numerical_columns)}")
  print(f"Categorical but Cardinal Columns: {len(categorical_but_cardinal)}")
  print(f"Numerical but Categorical Columns: {len(numerical_but_categorical)}")

  return categorical_columns, numerical_columns, categorical_but_cardinal, numerical_but_categorical
```

```python
# Feature Selection

# Variance Threshold Value

threshold = 0.01
selector = VarianceThreshold(threshold)
df_features = df[numerical_columns_x_date]
df_selected = selector.fit_transform(df_features)

# Selected Features
selected_features = df_features.columns[selector.get_support()]
print("Selected Features: ", selected_features)
```

```python
# Backward Elimination
'''
Backward Elimination is used to remove data from the dataset which is not that necessary than other data. So I think we can remove 'Adjusted Close' from our dataset.
'''

# Defining Features and target variable
X = df_features.drop('Adj Close', axis=1)
y = df_features['Adj Close']

# Building model
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())

# Removing features with high P-value
significant_level = 0.05
while True:
  max_p_value = max(model.pvalues)
  if max_p_value > significant_level:
    exclude_feature = model.pvalues.idxmax()
    X = X.drop(exclude_feature, axis=1)
    model = sm.OLS(y, X).fit()
  else:
    break

print("Selected Features: ", X.columns)
```

```python
# Recursive Feature Elimination (RFE)

'''
This Feature Selection method is used to select the feature which effetcts our model performance the best.
'''

# Building model and defining RFE object
model = LinearRegression()
rfe = RFE(model, n_features_to_select=3) # number of features to be selected

# Feature Selection with RFE
X = df_features.drop('Adj Close', axis=1)
y = df_features['Adj Close']
rfe = rfe.fit(X, y)

# Selected Features
selected_features_rfe = X.columns[rfe.support_]
print("Selected Features: ", selected_features_rfe)
```

```python
'''
As we can see that the values (graph) for The 'Volume' data is tilted to the right we can sey that the mean value is bigger than the median and the madian is bigger than the mode.
So, it possible to use the mean value for the outlier data but firstly we are going to calculate the Interquartile range of this data to get a clue of what the lower and upper limit of this box plot for
'Volume' is.
'''
```

```python
def calculate_outlier_thresholds(dataframe: pd.DataFrame, column_name: str, q1=0.25, q3=0.75):
  quartile_1 = dataframe[column_name].quantile(q1)
  quartile_3 = dataframe[column_name].quantile(q3)
  interquartile_range = quartile_3 - quartile_1
  upper_limit = (quartile_3 + 1.5) * interquartile_range
  lower_limit = (quartile_1 - 1.5) * interquartile_range

  return lower_limit, upper_limit
```

```python
calculate_outlier_thresholds(df, 'Volume')
```

```python
# Handling with outlier and extreme data using Winsorizing technique

df_winsorized = pd.DataFrame(df)

lower_limit = 4835280527572.5
upper_limit = 9761116761076.5

# Winsorizing process

df_winsorized['Volume_winsorized'] = np.where(df_winsorized['Volume'] < lower_limit, lower_limit, np.where(df_winsorized['Volume'] > upper_limit, upper_limit, df_winsorized['Volume']))

# Visualising new winsorized dataset
plt.figure(figsize=(20, 10))
plt.boxplot([df['Volume'], df_winsorized['Volume_winsorized']], vert=False, labels=['Original Volume', 'Winsorized Volume'])
plt.title("Stock Volumes - Before and After Winsorizing")
plt.grid(True)
plt.show()
```

```python
# Handling with outlier and extreme data using Z-Score technique

df_zscore = pd.DataFrame(df)

# Calculating Z-score
df_zscore['Volume_zscore'] = stats.zscore(df_zscore['Volume'])

# Filtering outlier data
threshold = 3
df_zscore['Volume_zscore_filtered'] = np.where(df_zscore['Volume_zscore'] > threshold, df_zscore['Volume'].mean() + threshold * df_zscore['Volume'].std(),
                                               np.where(df_zscore['Volume_zscore'] < -threshold, df_zscore['Volume'].mean() - threshold * df_zscore['Volume'].std(), df_zscore['Volume']))

# Visualising new Z-score dataset

plt.figure(figsize=(20, 10))
plt.boxplot([df['Volume'], df_zscore['Volume_zscore_filtered']], vert=False, labels=['Original Volume', 'Z-score Volume'])
plt.title("Stock Volumes - Before and After Z-score Calculation")
plt.grid(True)
plt.show()
```

```python
# Handling with outlier data using IQR technique

df_iqr = pd.DataFrame(df)

# Calculating IQR
quartile_1 = df_iqr['Volume'].quantile(0.25)
quartile_3 = df_iqr['Volume'].quantile(0.75)
interquartile_range = quartile_3 - quartile_1

# Outlier value limits
lower_bound = quartile_1 - 1.5 * interquartile_range
upper_bound = quartile_3 + 1.5 * interquartile_range

# Extreme value limits
# lower_bound_extreme = quartile_1 - 3 * interquartile_range
# upper_bound_extreme = quartile_3 + 3 * interquartile_range

# Filtering outlier values
df_iqr['Volume_iqr_filtered'] = np.where(df_iqr['Volume'] < lower_bound, lower_bound,
                                         np.where(df_iqr['Volume'] > upper_bound, upper_bound, df_iqr['Volume']))

# Visualising new IQR dataset

plt.figure(figsize=(20, 10))
plt.boxplot([df['Volume'], df_iqr['Volume_iqr_filtered']], vert=False, labels=['Original Volume', 'IQR Volume'])
plt.title("Stock Volumes - Before and After IQR Calculation")
plt.grid(True)
plt.show()
```
